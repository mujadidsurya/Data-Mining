## KESIMPULAN
Terdapat beberapa perbedaan antara PCA dan SVD. PCA didasarkan pada varians dalam dataset, sedangkan SVD didasarkan pada singular value dalam dataset. Perbedaan ini dapat menyebabkan perbedaan dalam interpretasi komponen utama atau komponen singular.
Selain itu, PCA lebih mudah untuk diimplementasikan daripada SVD. PCA dapat diimplementasikan menggunakan algoritma yang relatif sederhana, sedangkan SVD membutuhkan algoritma yang lebih kompleks.

Terdapat beberapa perbedaan antara SVD on the Digits Dataset - 3 Components dan PCA on the Digits Dataset - 3 Components. SVD on the Digits Dataset - 3 Components dapat mengekstrak lebih banyak informasi dari dataset daripada PCA on the Digits Dataset - 3 Components. Hal ini karena SVD on the Digits Dataset - 3 Components menggunakan tiga komponen singular, sedangkan PCA on the Digits Dataset - 3 Components hanya menggunakan tiga komponen utama.

TSNE membutuhkan lebih banyak waktu untuk diimplementasikan daripada PCA dan SVD. TSNE membutuhkan algoritma yang lebih kompleks untuk menghitung jarak antara poin dalam ruang tiga dimensi

# Transformasi Data Encoding dan Label Encoding

```python
# Membaca dataset yang dibutuhkan untuk dilakukan data encoding
df3 = pd.read_csv('Student_bucketing.csv')
df3.head()
```

## Data Encoding

```python
df_categorical = df3.select_dtypes(exclude=[np.number]) # Memilih variabel dengan tipe data kecuali variabel dengan tipe data numerik lalu disimpan pada df_categorical
df_categorical
```

Didapatkan bahwa dataset ini memiliki 2 variabel yang bertipe kategorik yaitu Grade dan Employed

```python
df3.Grade.value_counts()  # Melakukan perhitungan pada masing-masing nilai yang ada pada variabel Grade
```

Terdapat 2nd Class sebanyak 80, 3rd Class sebanyak 80, dan 1st Class sebanyak 72

```python
df3.Employed.value_counts() # Melakukan perhitungan pada masing-masing nilai yang ada pada variabel Employed
```

Terdapat nilai no sebanyak 133 dan yes sebanyak 99

```python
df3.Grade.replace({'1st Class':1, '2nd Class':2, '3rd Class':3}, inplace=True) # Melakukan encoding pada variabel Grade dengan fungsi replace
```

Menggunakan fungsi replace untuk melakukan data encoding yaitu mengubah data yang berbentuk kategorik menjadi numerik untuk mempermudah saat melakukan permodelan nanti
'1st Class' menjadi 1, '2nd Class' menjadi 2,dan  '3rd Class' menjadi 3

```python
df3.Employed.replace({'yes':1, 'no':0}, inplace=True)# Melakukan encoding pada variabel Employed dengan fungsi replace
```

Menggunakan fungsi replace untuk melakukan data encoding yaitu mengubah data yang berbentuk kategorik menjadi numerik untuk mempermudah saat melakukan permodelan nanti
'yes' menjadi 1 dan 'no' menjadi 0

```python
df3.head() # menampilkan hasil data yang telah dilakukan data encoding
```

Setiap nilai yang ada pada kolom Grade dan Employed sudah menjadi numerik

```python
df3.to_csv('data_transformation_encoding.csv')
```

## Label Encoding

```python
df4 = pd.read_csv('Banking_Marketing.csv')
df4.head()
```

```python
df4 = df4.dropna() # Menghapus nilai kosong pada dataset
```

```python
data_column_category =df4.select_dtypes(exclude=[np.number]).columns # Memilih variabel dengan tipe data kecuali variabel dengan tipe data numerik lalu disimpan pada df_categorical_column
data_column_category
```

Terdapat 10 kolom yang bernilai kategorik yaitu 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact','month', 'day_of_week', 'poutcome'

```python
df4[data_column_category].head() # Menampilkan data yang berisi kolom yang bertipe kategorik
```

```python
from sklearn.preprocessing import LabelEncoder # Memanggil fungsi LabelEncoder yang berasal dari library sklearn untuk melakukan label encoding

label_encoder = LabelEncoder() # Menggunakan fungsi LabelEncoder yang di definisikan ke label_encoder
# Melakukan perulangan untuk setiap variabel yang bernilai kategrik untuk dilakukan trasnsform dengan menggunakan fungsi LabelEncoder akan dilakupan pada setiap variabel yang bernilai kategorik
for i in data_column_category:
  df4[i] = label_encoder.fit_transform(df4[i])
print('Label Encoded Data: ')

df4.head()
```

```python
df4.to_csv('data_transformation_labeling_encoding.csv')
```
